```python
c=pd.get_dummies(train['HasCrCard'],drop_first=True)
train=pd.concat([train,c],axis=1)
train.head()
```

```python
train.rename(columns={1.0:'hascrcard'},inplace=True)
train.head()
```

```python
c=pd.get_dummies(train['IsActiveMember'],drop_first=True)
train=pd.concat([train,c],axis=1)
train.head()
```

```python
train.rename(columns={1.0:'activemember'},inplace=True)
train.head()
```

```python
train.drop(['Gender','HasCrCard','IsActiveMember'],axis=1,inplace=True)
```

```python
train.head()
```

```python
c=pd.get_dummies(train['NumOfProducts'],drop_first=True)
train=pd.concat([train,c],axis=1)
train.head()
```

```python
train.rename(columns={2:'2-Products',3:'3-Products',4:'4-Prducts'},inplace=True)
train.head()
```

```python
train['Male'] = train['Male'].astype(int)
train.head()
```

```python
train['hascrcard'] = train['hascrcard'].astype(int)
train['activemember'] = train['activemember'].astype(int)
train['2-Products'] = train['2-Products'].astype(int)
train['3-Products'] = train['3-Products'].astype(int)
train.head()
```

```python
train['4-Prducts'] = train['4-Prducts'].astype(int)
train.head()
```

```python
c=pd.get_dummies(train['Geography'],drop_first=True)
train=pd.concat([train,c],axis=1)
train.head()
```

```python
train['Germany'] = train['Germany'].astype(int)
train['Spain'] = train['Spain'].astype(int)
train.head()
```

```python
train.drop('Surname',inplace=True,axis=1)
train.head()
```

```python
train.drop('NumOfProducts',inplace=True,axis=1)
train.drop('Geography',inplace=True,axis=1)
train.head()
```

```python
X_train = train.drop('Exited',axis=1)
y_train = train['Exited']
```

```python
logmodel.fit(X_train,y_train)
```
